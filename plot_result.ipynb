```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
fig = plt.figure(figsize=(6, 5), dpi=600)
sns.set_theme(style='whitegrid', font_scale=0.85, rc={'lines.linewidth': 1.5})
palette = sns.color_palette("tab10")

df = pd.read_csv('files/data.csv')

eps_num_ppo = df.iloc[:, 0].to_numpy()
cost_ppo = df.iloc[:, 1].to_numpy()
ret_ppo = df.iloc[:, 2].to_numpy()

eps_num_ppol = df.iloc[:, 3].to_numpy()
cost_ppol = df.iloc[:, 4].to_numpy()
ret_ppol = df.iloc[:, 5].to_numpy()

plt.subplot(2,2,1)
plt.plot(eps_num_ppol, ret_ppol, color=palette[2])
plt.ylim(0, 260)
plt.ylabel('Episodic Return')
plt.gca().set_xticklabels([])
plt.title("PPO-Lagrangian")

plt.subplot(2,2,2)
plt.plot(eps_num_ppo, ret_ppo, color=palette[1])
plt.ylim(0, 260)
plt.gca().set_xticklabels([])
plt.title("PPO")

plt.subplot(2,2,3)
plt.plot(eps_num_ppol, cost_ppol, color=palette[2])
plt.ylim(-100, 2500)
plt.ylabel('Episodic Cost')
plt.xlabel('Number of Episodes')

plt.subplot(2,2,4)
plt.plot(eps_num_ppo, cost_ppo, color=palette[1])
plt.ylim(-100, 2500)
plt.xlabel('Number of Episodes')

plt.show()
```
